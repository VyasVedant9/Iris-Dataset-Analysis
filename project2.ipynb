# Section 4: Data Science

### 4.1: Correlation

The question I have about my dataset is how closely correlated are the factors such as the petal width, petal length, sepal width and the sepal length. In order to understand the correlation between these different factors, I will use the inbuilt function of corr which will give me an output of a correlation matrix. In order to use the corr function I will first filter it to contain only numerical columns. I will interpret the correlation matrix to make inferences about the correlation of the factors stated above. 

```python
numerical_cols = ["Id",	"SepalLengthCm","SepalWidthCm",	"PetalLengthCm"	,"PetalWidthCm"]
df[numerical_cols].corr()
```

These are the main observations I have from the correlation matrix:
1. The petal width and sepal length has a strong positive correlation with a correlation coefficient of 0.817954
2. The petal width and petal length has a strong positive correlation with a correlation coefficient of 0.962757
3. The petal length and sepal length has a strong positive correlation with a correlation coefficient of 0.871754

### 4.2: Linear Regression Model

I will next work on a linear regression model which will have the petal length as the independent variable and the sepal length as the dependent variable. 

```python
from sklearn.linear_model import LinearRegression
df_num = df[numerical_cols]
model = LinearRegression()
model.fit(df_num[["PetalLengthCm"]], df_num["SepalLengthCm"])
df_num["Sepal Length Prediction"] = model.predict(df_num[["PetalLengthCm"]])
df_num["Sepal Length Residue"] = df_num["SepalLengthCm"] - df_num["Sepal Length Prediction"]
df_num
```

The next linear regression model is a model which will use the petal length as the independent variable and the petal width as the dependent variable. 

```python
model2 = LinearRegression()
model2.fit(df_num[["PetalLengthCm"]], df_num["PetalWidthCm"])
df_num["Petal Width Prediction"] = model2.predict(df_num[["PetalLengthCm"]])
df_num["Petal Width Residue"] = df_num["PetalWidthCm"] - df_num["Petal Width Prediction"]
df_num
```
